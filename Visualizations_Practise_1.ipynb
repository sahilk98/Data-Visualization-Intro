```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

USING A GLOBAL API

```
x = np.arange(-10,11)
```

```
plt.figure(figsize = (12,6))
plt.title("Practise Plot")
plt.plot(x,x**2); plt.plot(x, -(x**2))
```

```
plt.figure(figsize = (15,8))
plt.title("Multiple subplot shit")

#subplot 1
plt.subplot(1,2,1)  # Here, the first parameter is thr row, then the column, and the last one is the index, as in which one is being edited
# using this particular piece of code.
plt.plot(x,x+2)
plt.plot([0,10,20], [-10,0,10]) # Basically, (start point, end point, y intercept)

#subplot 2
plt.subplot(1,2,2)
plt.plot(x,x+3)
plt.plot([10,0,10], [0,3,0])

#subplot 3
#plt.subplot(1,3,3)
#plt.plot(x, 2*x)
```

Object Oriented Approach for visualizations

```
#fig, axis = plt.subplots(figsize =(12,6))
```

```
plot_objects = plt.subplots(nrows = 2, ncols=2, figsize=(16,10))

```

```
fig, ((ax1, ax2), (ax3, ax4)) = plot_objects
```

```
ax1.plot(np.random.randn(25), c = 'yellow')   #plotting random points on the graph
ax2.plot(np.random.randn(25))
ax3.plot(np.random.randn(25))
ax4.plot(np.random.randn(25))
fig

# With an object oriented approach, the edits can be done by selecting
# the specific axis and then making the changes right there from anywhere
# in the code. Simpler than the global variable approach.
```
